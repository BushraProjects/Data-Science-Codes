```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
diabete = pd.read_csv("diabetes.csv")
```

```python
x = diabete.drop('Outcome', axis = 1)
y = diabete['Outcome']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
logModel = LogisticRegression()
logModel.fit(x_train, y_train)
```

```python
predictions = logModel.predict(x_test)
classification_report(y_test, predictions)
```

```python
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import accuracy_score
from sklearn import preprocessing

accuracy_score(y_test, predictions)
```

```python
lengthData = np.arange(0, 231, 1).tolist()

plt.scatter(lengthData, y_test, label = 'Actual', color = 'k', marker = "o")
plt.scatter(lengthData, predictions, label = 'prediction', color='r', marker="*")
plt.legend()
plt.xlim(0, 231)
plt.show()
```

```python
plt.plot(x.BMI)
plt.plot(x.BloodPressure)
plt.show
```

```python
# Training for the second time

scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(x)
scaled_df = pd.DataFrame(scaled_df)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_df, y, test_size = 0.3, random_state = 1)
logmodelN = LogisticRegression()
logmodelN.fit(x_train, y_train)
```

```python
predictionsN = logmodelN.predict(x_test)
classification_report(y_test, predictionsN)
```

```python
accuracy_score(y_test, predictionsN)
```
